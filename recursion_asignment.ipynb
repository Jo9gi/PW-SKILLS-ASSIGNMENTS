**Q.1 Can you explain the logic and working of the Tower of Hanoi algorithm by writing a Java program? How does the recursion work, and how are the movements of disks between rods accomplished?**

Screenshot 2024-03-27 101550.png

```python
def tower_of_hanoi(num, source_rod, destination_rod, auxiliary_rod):
    if num == 1:
        print(f"Move disk 1 from rod {source_rod} to rod {destination_rod}")
        return

    tower_of_hanoi(num - 1, source_rod, auxiliary_rod, destination_rod)
    print(f"Move disk {num} from rod {source_rod} to rod {destination_rod}")
    tower_of_hanoi(num - 1, auxiliary_rod, destination_rod, source_rod)

if __name__ == "__main__":
    num = 3
    source_rod = 'A'
    destination_rod = 'C'
    auxiliary_rod = 'B'

    print(f"Steps to solve Tower of Hanoi with {num} disks:")
    tower_of_hanoi(num, source_rod, destination_rod, auxiliary_rod)
```

**Q. 3 Print the max value of the array [ 13, 1, -3, 22, 5].**

```python
def maximum(arr):
  #base_condition
  if len(arr)==1:
    return arr[0]
  else:
    val = maximum(arr[1:])
    return arr[0] if arr[0]>val else val
#driver code
arr=[ 13, 1, -3, 22, 5]
maximum(arr)


##the time complexity is O(n)
##the space complexity is O(n)
```

**Q.5 Given a number n. Print if it is an armstrong number or not.An armstrong number is a number if the sum of every digit in that number raised to the power of total digits in that number is equal to the number.**

Example : 153 = 1^3 + 5^3 + 3^3 = 1 + 125 + 27 = 153 hence 153 is an armstrong number. (Easy)

Input1 : 153

Output1 : Yes

Input 2 : 134

Output2 : No

```python
def armstrong_number(number,digits=None):
  if digits is None:
    digits = len(str(number))
#base condition
  if number==0:
    return 0
#driver code
  else:
    num = number % 10
    power = num**digits
    return power+armstrong_number(number//10,digits)

input1 = 153
result = armstrong_number(input1)
if result == input1:
  print("Output1 :Yes")
else:
  print("Output1 :No")

input2 = 134
result = armstrong_number(input2)
if result == input2:
  print("Output2 :Yes")
else:
  print("Output2 :No")
```
